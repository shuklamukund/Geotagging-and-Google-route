```python
import pandas as pd
import numpy as np
import requests
```

```python
df = pd.read_excel('location_mukund.xlsx')
```

```python
df = pd.DataFrame(df['location'].apply(lambda x: x.strip()))
```

```python
df.drop_duplicates(subset=['location'], keep='first', inplace=True)
```

```python
df = pd.DataFrame(df['location'].apply(lambda x: x.strip().replace(" ","+")))
```

```python
def get_data(data1) :
    url1 = "https://nominatim.openstreetmap.org/search.php?q="
    url2=str(data1)
    url3 = '&format=jsonv2'

    url = url1 + url2 + url3
    response = requests.get(url, verify=False)
    return response.json()
```

```python
dfs=[]

for i, r in df.iterrows():
    try:
        data1=r['location']
        results=get_data(data1)
        temp_df=pd.DataFrame(results)
        dfs.append(temp_df)
    except:
        pass


finaldf=pd.concat(dfs)
```
